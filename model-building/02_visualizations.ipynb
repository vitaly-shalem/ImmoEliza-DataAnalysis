```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#   load data
df = pd.read_csv("..\\data\\03_ml_prep_data_drop_duplicate.csv", index_col="ID")
df.info()
```

```python
dfCorr = df[[
    "typeNum", "regionNum", "provinceNum", 
    "postalCode", "localityType",
    "bedroomCount", "netHabitableSurface", "conditionNum", 
    "price"]].corr()
sns.heatmap(dfCorr, annot=True)
plt.show()
```
